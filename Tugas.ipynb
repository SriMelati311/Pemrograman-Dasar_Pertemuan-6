```python
import datetime

# Data menu menggunakan dictionary (non-primitif)
menu_kafe = {
    1: ("Jus Mangga", 18000),
    2: ("Jus Jeruk", 18000),
    3: ("Jus Alpukat", 20000),
    4: ("Jus Sirsak", 19000)
}

# Fungsi menampilkan menu
def tampilkan_menu():
    print("="*40)
    print("SELAMAT DATANG DI JUSSS BORLAB")
    print("="*40)
    print("Menu Makanan & Minuman")
    for key, value in menu_kafe.items():
        print(f"{key}. {value[0]:<12} - Rp {value[1]}")
    print("-"*40)

# Fungsi menghitung total dan diskon
def hitung_total(pesanan):
    total = 0
    for item, jumlah in pesanan.items():
        total += menu_kafe[item][1] * jumlah
    return total

# Fungsi menghitung diskon berdasarkan total
def hitung_diskon(total):
    if total > 50000:
        return 0.70
    elif total > 30000:
        return 0.50
    elif total > 20000:
        return 0.30
    else:
        return 0

# Fungsi mencetak struk pembayaran
def cetak_struk(pesanan, total, diskon):
    potongan = int(total * diskon)
    total_bayar = total - potongan

    print("\n" + "="*40)
    print("         STRUK PEMBAYARAN")
    print("Tanggal :", datetime.datetime.now().strftime("%d-%m-%Y %H:%M:%S"))
    print("="*40)
    print(f"{'Nama Item':<15}{'Qty':<5}{'Harga':<10}{'Subtotal'}")
    print("-"*40)

    # Looping untuk menampilkan pesanan
    for item, jumlah in pesanan.items():
        if jumlah > 0:
            nama, harga = menu_kafe[item]
            subtotal = harga * jumlah
            print(f"{nama:<15}{jumlah:<5}Rp {harga:<8}Rp {subtotal}")

    print("-"*40)
    print(f"{'Total Harga':<16}: Rp {total}")
    print(f"{'Diskon':<16}: {int(diskon*100)}% (Rp {potongan})")
    print(f"{'Total Bayar':<16}: Rp {int(total_bayar)}")
    print("="*40)

    # Proses pembayaran dengan percabangan
    while True:
        uang_dibayar = int(input("Uang Pembayaran : Rp "))
        if uang_dibayar >= total_bayar:
            kembalian = uang_dibayar - total_bayar
            print(f"{'Kembalian':<16}: Rp {kembalian}")
            print("="*40)
            print("Terima kasih telah berbelanja di JUSSS BORLAB!")
            print("Semoga hari Anda menyenangkan :)")
            break
        else:
            kekurangan = total_bayar - uang_dibayar
            print(f"Uang Anda kurang Rp {kekurangan}. Silakan masukkan lagi.")

# Fungsi utama program
def main():
    tampilkan_menu()

    pesanan = {}  # Dictionary untuk menyimpan pesanan
    pilihan_ulang = True  # Boolean sebagai kontrol perulangan

    while pilihan_ulang:
        try:
            kode = int(input("Masukkan nomor menu (1-4): "))
            if kode in menu_kafe:
                jumlah = int(input(f"Masukkan jumlah {menu_kafe[kode][0]}: "))
                pesanan[kode] = pesanan.get(kode, 0) + jumlah
            else:
                print("Menu tidak tersedia!")

            lanjut = input("Tambah pesanan lain? (y/n): ").lower()
            if lanjut == "n":
                pilihan_ulang = False
        except ValueError:
            print("Input tidak valid! Harap masukkan angka.")

    total = hitung_total(pesanan)
    diskon = hitung_diskon(total)
    cetak_struk(pesanan, total, diskon)

# Jalankan program
main()
```
